```python
# import modules
import pandas as pd
```

```python
# import metadata
md = pd.read_csv("../../data/srr_numbers_with_metadata_12-May-2023.csv",sep=",",header="infer")
md.head()
```

```python
# replace BioProject Brandon_Webster with PRJNA904734
md.loc[md["BioProject"]=="Brandon_Webster","BioProject"]="PRJNA904734"
```

```python
# how many unique genotypes are there right now?
len(md["Genotype"].unique())
```

```python
# on to the genotypes: start with the hybrids
for i in list(md["Genotype"].unique()):
    if "x" in i:
        print(i)
    if "X" in i:
        print(i)
```

Make it so that all hybrids are designated by a lowercase x. Collapse the duplicate B73xMo17s.

```python
# replace values
md.loc[md["Genotype"]=="B73XMO17","Genotype"]="B73xMo17"
md.loc[md["Genotype"]=="B73XPHN82","Genotype"]="B73xPHN82"
md.loc[md["Genotype"]=="B73XPHZ51","Genotype"]="B73xPHZ51"
md.loc[md["Genotype"]=="LH195XPHN82","Genotype"]="LH195xPHN82"
md.loc[md["Genotype"]=="PHB47XPHN82","Genotype"]="PHB47xPHN82"
```

```python
# check number of unique genotypes
len(md["Genotype"].unique())
```

```python
# make a list of the corrected hybrids
corrhy = ["B73xMo17","B73xPHN82","B73xPHZ51","LH195xPHN82","PHB47xPHN82"]
```

```python
# identify other duplicated genotypes & correct them
for i in sorted(list(md["Genotype"].unique())):
    if i not in corrhy:
        print(i)
```
