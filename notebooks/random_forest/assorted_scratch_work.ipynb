## Set up the TPM file for WGCNA

```python
## drop BioProject & Treatment columns
tpmd = tpm.drop(["BioProject","Treatment"],axis=1)
tpmd.head()
```

```python
tpmt.to_csv("../../data/rawtpm_forWGCNA_noPEG.tsv",sep="\t",header=True,index=False)
```

```python
# load metadata
md = pd.read_csv("../../data/srr_numbers_with_metadata_12-May-2023.csv",sep=",",header="infer",usecols=["Sample","BioProject","Tissue"])
md.head()
```

```python
tpmtissue = md.merge(tpm)
tpmtissue.head()
```

## Data exploration: stressor representation, numbers of BioProjects, etc.

```python
# subset to just leaf tissue and find out if all stressors are represented
tpmtissue[tpmtissue["Tissue"]=="Leaf"]["Treatment"].unique()
```

```python
leaf = tpmtissue[tpmtissue["Tissue"]=="Leaf"].drop("Tissue",axis=1)
```

```python
leaf.head()
```

```python
# count number of BioProjects for each stressor when including all tissues
for s in ["Drought","Salt","Cold","Heat","Low_Nitrogen","Flooding"]:
    stpm = tpmtissue[tpmtissue["Treatment"]==s]
    print(s)
    print(len(stpm["BioProject"].unique()))
```

```python
# now do it again for including only leaf tissue
for s in ["Drought","Salt","Cold","Heat","Low_Nitrogen","Flooding"]:
    stpm = leaf[leaf["Treatment"]==s]
    print(s)
    print(len(stpm["BioProject"].unique()))
```

```python
tpmtissue["Tissue"].unique()
```

```python
vegtissues = ["Leaf","Root","Shoot","Stalk","Leaf meristem"]
reptissues = ["Ear","Kernel","Tassel","Silk","Pollen","Embryo","Ovary"]
```

```python
veg = tpmtissue[tpmtissue["Tissue"].isin(vegtissues)].drop("Tissue",axis=1)
rep = tpmtissue[tpmtissue["Tissue"].isin(reptissues)].drop("Tissue",axis=1)
```

```python
# count BioProjects per stressor in vegetative tissues
for s in ["Drought","Salt","Cold","Heat","Low_Nitrogen","Flooding"]:
    stpm = veg[veg["Treatment"]==s]
    print(s)
    print(len(stpm["BioProject"].unique()))
```

```python
# count BioProjects per stressor in reproductive tissues
for s in ["Drought","Salt","Cold","Heat","Low_Nitrogen","Flooding"]:
    stpm = rep[rep["Treatment"]==s]
    print(s)
    print(len(stpm["BioProject"].unique()))
```

```python
# save leaf, veg, and rep
leaf.to_csv("../../data/rawtpm_leafonly.tsv",sep="\t",header=True,index=False)
veg.to_csv("../../data/rawtpm_vegtissues.tsv",sep="\t",header=True,index=False)
rep.to_csv("../../data/rawtpm_reptissues.tsv",sep="\t",header=True,index=False)
```

```python
# find out how many samples for each dataset
for s in ["Drought","Salt","Cold","Heat","Low_Nitrogen","Flooding"]:
    if s=="Drought":
        stpm = tpm[tpm["Treatment"].isin(["Drought","DroughtRepeat"])]
    else:
        stpm = tpm[tpm["Treatment"]==s]
    print("Number of samples for stressor "+s+": "+str(len(stpm["Sample"].unique())))
```

```python
for s in ["Drought","Salt","Cold","Heat","Low_Nitrogen","Flooding"]:
    if s=="Drought":
        stpm = tpm[tpm["Treatment"].isin(["Drought","DroughtRepeat"])]
    else:
        stpm = tpm[tpm["Treatment"]==s]
    sbp = stpm["BioProject"].unique()
    sbptpm = tpm[tpm["BioProject"].isin(sbp)]
    sbptpm = sbptpm[sbptpm["Treatment"].isin(["Control",s])]
    print("Number of samples for stressor "+s+" + controls: "+str(len(sbptpm["Sample"].unique())))
```
