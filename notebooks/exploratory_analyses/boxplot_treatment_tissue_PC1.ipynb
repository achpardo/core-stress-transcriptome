```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
expmd = pd.read_csv('../../data/principal_components_metadata.csv')
```

```python
sns.boxplot(x=expmd["Treatment"],y=expmd["PC1"])
```

```python
sns.boxplot(y=expmd["Tissue"],x=expmd["PC1"])
```

```python
# how many genotypes with >20 samples?
gt20 = []
for g in list(expmd["Genotype"].unique()):
    df = expmd[expmd["Genotype"]==g]
    if len(df.index)>20:
        gt20.append(g)
```

```python
gt20
```

```python
dfgt20 = expmd[expmd["Genotype"].isin(gt20)]
len(dfgt20.index)
```

```python
# save dfgt20
dfgt20.to_csv("../../data/principal_components_metadata_gt20samp.csv",sep=",",header=True,index=False)
```
